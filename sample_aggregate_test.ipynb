```python
# This is my brute-force method using C++ style iterators and loops
test_df['m2'] = None
test_df
```

```python
it = iter(test_df.groupby('item')['measure'])

idx = 0
for i in it:
    last_x = None
    s = f"{i[0]}: "
    n = 0
    for x in i[1]:
        s = f"{s} {x}"

        if n == 0:
            # first row of the group is always False
            test_df.loc[idx, 'm2'] = False
        else:
            # now do the change test
            test_df.loc[idx, 'm2'] = (x != last_x)

        # save this one for the next loop
        last_x = x
        n = n + 1
        idx = idx + 1

    if idx < 30: print(s)

test_df
```
